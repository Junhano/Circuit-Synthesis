```python
def parseCadenceData(files, output_variable_names):
    import pandas as pd
    import numpy as np
    import re
    
    assert len(files) > 0, 'Make sure there are at least one file'
    assert len(files) == len(output_variable_names)
    
    column_index_dict = dict()
    row_index_dict = dict()
    
    file_df = pd.read_csv(files[0])
    
    pattern = r'"([a-zA-Z0-9]+)"[\s]*((?:[0-9]*\.)?[0-9]+e?-?[0-9]+)'
    
    for i in range(0, len(file_df.columns) - 1):
        analyze_string = file_df.columns[i + 1]
        matching = re.findall(pattern,analyze_string)
        key_array = []
        for z in matching:
            key_array.append(z)
        key_array.sort(key=lambda x: x[0])
        column_index_dict[i] = tuple(key_array)
    
    for i in range(len(file_df[file_df.columns[0]])):
        row_index_dict[i] = file_df[file_df.columns[0]].iloc[i]
        
    coordinate_dict = dict()
    index_name = None
    
    for index, file in enumerate(files):
        temp_df = pd.read_csv(file)
        
        if index_name is None:
            index_name = temp_df.columns[0].strip()
        
        for i in range(len(temp_df)):
            for c in range(len(temp_df.iloc[i]) - 1):

                temp_val = temp_df.iloc[i, c+1]
                if (i,c) not in coordinate_dict.keys():
                    coordinate_dict[(i,c)] = [temp_val]
                else:
                    coordinate_dict[(i,c)].append(temp_val)

    
    
    row_data_point = []
    output_column_names = []
    
    output_column_names.append(index_name)
    for value_tuples in column_index_dict.values():
        for t in value_tuples:
            output_column_names.append(t[0])
        break
    output_column_names += output_variable_names
    
    for row in range(max(row_index_dict.keys()) + 1):
        for column in range(max(column_index_dict.keys()) + 1):
            temp_data = list()
            temp_data.append(row_index_dict[row])
            for t in column_index_dict[column]:
                temp_data.append(t[1])
            for v in coordinate_dict[(row, column)]:
                temp_data.append(round(v,7))
            row_data_point.append(temp_data)
        
    row_data_point = np.array(row_data_point)
    
    row_data_df = pd.DataFrame(row_data_point, columns = output_column_names)
    
    row_data_df.to_csv("./combine.csv", index=False)
```

```python
k = parseCadenceData(["./Data/Power.csv", "./Data/Power.csv"], ["power", "power 2"])
k.head()
```
