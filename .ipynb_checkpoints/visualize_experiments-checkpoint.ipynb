```python
def rand_jitter(arr):
    stdev = .01 * (max(arr) - min(arr))
    return arr + np.random.randn(len(arr)) * stdev

def jitter(x, y, s=None, c='b', marker='o', cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, verts=None, hold=None, **kwargs):
    return plt.scatter(rand_jitter(x), rand_jitter(y), s=s, c=c, marker=marker, cmap=cmap, norm=norm, vmin=vmin, vmax=vmax, alpha=alpha, linewidths=linewidths, **kwargs)
```

## Load Data

"Accuracy" is training accuracy, validation accuracy is validation accuracy

```python
experiment_name = "Feb-17-best-model"
results = utils.openPickle(f"Experiments/{experiment_name}")
print("Avaliable test metrics")
print(results.columns)
```

```python
acc_optim = results.loc[:,["accuracy","config.optim"]]
optim_str = acc_optim.iloc[:,1].astype(str)
unique = optim_str.unique()
optim_dict = dict()
optim = np.array(optim_str)
for i,o in enumerate(unique):
    idx = np.where(optim==unique[i])
    optim[idx] = i
    print(i,o)
fig, ax = plt.subplots()
ax.set_xticks(range(len(unique)))
ax.set_xticklabels(unique, rotation = 90)
plt.xlabel("optimizer")
plt.ylabel("accuracy")
jitter(optim,results.loc[:,"accuracy"], marker = "." )
```

```python
acc_optim = results.loc[:,["accuracy","config.loss_fn"]]
optim_str = acc_optim.iloc[:,1].astype(str)
unique = optim_str.unique()
optim_dict = dict()
optim = np.array(optim_str)
for i,o in enumerate(unique):
    idx = np.where(optim==unique[i])
    optim[idx] = i
    print(i,o)
fig, ax = plt.subplots()
ax.set_xticks(range(len(unique)))
ax.set_xticklabels(unique, rotation = 90)
plt.xlabel("Loss_Finction")
plt.ylabel("accuracy")
plt.scatter(optim,results.loc[:,"accuracy"], marker = "." )
```

On the plots with labels I call jitter which randomly adds noise to the x value of each pixel to avoid data points being hidden behind one another. All pixels in a vertical group belong to the same label. The slight change in x values are meaningless.
